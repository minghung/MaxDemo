```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
stock1 = pd.read_csv("2303.csv")
stock2 = pd.read_csv("2330.csv")
```

```python
import datetime
```

```python
def toWesternDate(taiwanDate):
    """taiwanDate: YYY/MM/DD"""
    year = int(taiwanDate[:taiwanDate.index("/")])+1911
    new_date = str(year)+taiwanDate[taiwanDate.index("/"):]
    return datetime.datetime.strptime(new_date, "%Y/%m/%d")
```

```python
stock1.dropna(inplace=True)
```

```python
stock1["newDate"] = stock1.apply(lambda row: toWesternDate(row["date"]), axis=1)
```

```python
stock2.dropna(inplace=True)
```

```python
stock2["newDate"] = stock1.apply(lambda row: toWesternDate(row["date"]), axis=1)
```

```python
stock1.rename(columns={"2303":"price"}, inplace=True)
stock2.rename(columns={"2330":"price"}, inplace=True)
```

```python
stock2.head()
```

```python
plt.plot(stock1["newDate"], stock1["newPrice"])
plt.plot(stock2["newDate"], stock2["newPrice"])
plt.legend(["2303","2330"])
plt.show()
```
